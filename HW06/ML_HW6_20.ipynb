```python
import os
import shutil

path = '/content/40/StyleGAN2/'
new_path = '/content/submission_20/'

if os.path.isdir(new_path) == False:
    os.mkdir(new_path)

for _image in os.listdir(path):
    if _image[-6:] != 'mr.jpg' and _image[-4:] != '.jpg':
        shutil.move(os.path.join(path, _image), os.path.join(new_path, _image))

for idx, _image in enumerate(os.listdir(new_path)):
    os.rename(os.path.join(new_path, _image), os.path.join(new_path, str(idx+1)+'.jpg'))
```
